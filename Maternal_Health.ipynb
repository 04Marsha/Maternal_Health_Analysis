```python
import pandas as pd

df = pd.read_csv('cleaned_maternal_health_data.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x=df['sector'])
plt.xticks(rotation=85)
plt.title("Data by Sector")
plt.show()
```

```python
indicator_df = df[df['indicator'] == '3.1.1: Maternal Mortality Ratio (MMR), (per 1,00,000 live births)']
indicator_df = indicator_df.sort_values('timeperiod')

import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(indicator_df['timeperiod'], indicator_df['datavalue'], marker='o', linestyle='-')
plt.title('Maternal Mortality Ratio in India Over Time')
plt.xlabel('Year')
plt.ylabel('MMR (per 100,000 live births)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
mmr_df = df[df['indicator'].str.contains('Maternal Mortality Ratio', na=False)]
print(mmr_df[['timeperiod', 'datavalue']])
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

X = mmr_df[['Year']]
y = mmr_df['MMR']

model = LinearRegression()
model.fit(X, y)

future_years = pd.DataFrame({'Year': range(2015, 2031)})
future_mmr = model.predict(future_years)

plt.figure(figsize=(10, 6))
plt.plot(mmr_df['Year'], mmr_df['MMR'], 'bo-', label='Actual MMR')
plt.plot(future_years['Year'], future_mmr, 'r--', label='Forecasted MMR')
plt.axhline(70, color='green', linestyle=':', label='SDG Target (70)')
plt.xlabel('Year')
plt.ylabel('MMR')
plt.title('Maternal Mortality Ratio Forecast (India)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

X = mmr_df['Year'].values.reshape(-1, 1)
y = mmr_df['MMR'].values
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
future_years = np.arange(2018, 2031).reshape(-1, 1)
future_predictions = model.predict(future_years)
```

```python
plt.figure(figsize=(10, 6))

plt.plot(mmr_df['Year'], mmr_df['MMR'], marker='o', label='Actual MMR')

plt.plot(future_years, future_predictions, marker='x', linestyle='--', label='Forecasted MMR', color='green')

plt.title('MMR Forecast up to 2030')
plt.xlabel('Year')
plt.ylabel('MMR (per 100,000 live births)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
